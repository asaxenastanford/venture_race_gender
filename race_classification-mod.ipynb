```python
bc = BertClient(check_length=False)
```

```python
print("Training data")
book = xlrd.open_workbook("race_training_binary.xlsx")
sheet = book.sheet_by_index(0)

X_str_train = []
y_train = []

for row in range(2,30162):
    first = sheet.cell_value(row,1)
    first = first.lower()
    last = sheet.cell_value(row,2)
    last = last.lower()
    name = first + " " + last
    race = sheet.cell_value(row,19)
    race = race.lower()
    X_str_train.append(name)
    y_train.append(race)
    if(row % 10 == 0):
    	print(row)
    	print(name)
    	print(race)

print(X_str_train)
print(y_train)
```

```python
print(len(X_str_train))
print(len(y_train))
```

```python
print("Training data")
booktraining = xlrd.open_workbook("race_training_binary_reduced.xlsx")
sheettraining = booktraining.sheet_by_index(0)

X_str_train_red = []
y_train_red = []

for row in range(2,24127):
    first = sheettraining.cell_value(row,1)
    first = first.lower()
    last = sheettraining.cell_value(row,2)
    last = last.lower()
    name = first + " " + last
    race = sheettraining.cell_value(row,19)
    race = race.lower()
    X_str_train_red.append(name)
    y_train_red.append(race)
    if(row % 10 == 0):
    	print(row)
    	print(name)
    	print(race)

print(X_str_train_red)
print(y_train_red)
```

```python
print(len(X_str_train_red))
print(len(y_train_red))
```

```python
book1 = xlrd.open_workbook("race_test.xlsx")
testsheet = book1.sheet_by_index(0)

X_str_dev = []
y_dev = []

for row in range(1,1317):
    first = testsheet.cell_value(row,0)
    first = first.lower()
    last = testsheet.cell_value(row,1)
    last = last.lower()
    name = first + " " + last
    X_str_dev.append(name)
    race = testsheet.cell_value(row,7)
    race = race.lower()
    y_dev.append(race)
    
print(X_str_dev)
print(y_dev)
print(len(X_str_dev))
print(len(y_dev))
```

```python
book2 = xlrd.open_workbook("race_testing_binary_reduced.xlsx")
testsheetred = book2.sheet_by_index(0)

X_str_dev_red = []
y_dev_red = []

for row in range(0,6035):
    first = testsheetred.cell_value(row,1)
    first = first.lower()
    last = testsheetred.cell_value(row,2)
    last = last.lower()
    name = first + " " + last
    X_str_dev_red.append(name)
    race = testsheetred.cell_value(row,19)
    race = race.lower()
    y_dev_red.append(race)
    
print(X_str_dev_red)
print(y_dev_red)
print(len(X_str_dev_red))
print(len(y_dev_red))
```

```python
def bert_reduce_mean(X):
    return X.mean(axis=1)
```

```python
X_bert_train_mean = bert_reduce_mean(X_bert_train)
X_bert_train_mean_red = bert_reduce_mean(X_bert_train_red)
```

```python
X_bert_train_mean.shape[1]
```

```python
X_bert_train_mean_red.shape[1]
```

```python
mod = TorchShallowNeuralClassifier(
    max_iter=100, hidden_dim=300)

mod_red = TorchShallowNeuralClassifier(
    max_iter=100, hidden_dim=300)
```

```python
%time _ = mod.fit(X_bert_train_mean, y_train)
```

```python
%time _ = mod_red.fit(X_bert_train_mean_red, y_train_red)
```

```python
X_bert_dev_mean = bert_reduce_mean(X_bert_dev)
```

```python
X_bert_dev_mean_red = bert_reduce_mean(X_bert_dev_red)
```

```python
bert_sent_preds = mod.predict(X_bert_dev_mean)

import openpyxl
xfile1 = openpyxl.load_workbook('test_results.xlsx')
javasheetwrite = xfile1['Sheet1'] 

counter = 1

for race in bert_sent_preds:
    index = "A" + str(counter)
    javasheetwrite[index] = race
    counter += 1

xfile1.save('test_results.xlsx')
    
```

```python
bert_sent_preds_red = mod_red.predict(X_bert_dev_mean_red)

counter = 1

for race in bert_sent_preds_red:
    index = "B" + str(counter)
    javasheetwrite[index] = race
    counter += 1

xfile1.save('test_results.xlsx')
```

```python
print(classification_report(y_dev, bert_sent_preds, digits=3))
```
